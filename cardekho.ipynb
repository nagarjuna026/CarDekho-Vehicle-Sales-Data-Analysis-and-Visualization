```python
#Which is the most cost depreciated vehicle in data ?
df['depriciation'] = df['Present_Price'] - df['Selling_Price']
df.loc[df.depriciation == df.depriciation.max()]
```

```python
#Which is the least cost depreciated vehicle in data ?
df.loc[df.depriciation == df.depriciation.min()]
```

```python
#Which brands of vehicles are less affected by cost depreciation
df['Depreciation'] = (df['Present_Price'] - df['Selling_Price']) / df['Present_Price'] * 100
average_depreciation_by_brand = df.groupby('Car_Name')['Depreciation'].mean()
least_depreciation_brands = average_depreciation_by_brand.sort_values()
least_depreciation_brands
```
